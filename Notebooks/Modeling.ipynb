# Song Genres Classifier Function

Product Use Case: Music streaming platforms like Spotify might not be able to label all songs that come out on a hourly basis on their platform into their respective genres. Therfore, using a ML model to predict the genre(s) the song might belong to would be useful for the company. The function below classifies the song into one of 25 genre labels and in addition gives the next 2 most likely genre labels that it could be potentially classified into, in case the company decided to classify songs into more than one genre if the probabilty of classification into those other genres reaches a certain thershold of their choosing.

```python
# Request access token

def get_token():

    client_id = 'd034330dcdd940e3ab87215e7fe86d9d'
    client_secret = '5e5a5f7c06a74cbdbfe64051225e7943'
    grant_type = 'client_credentials'
    body_params = {'grant_type': grant_type}

    url = 'https://accounts.spotify.com/api/token'

    response = requests.post(url, data = body_params, auth = (client_id, client_secret)) 
    token = response.json()['access_token']
    return token 
```

```python
# Search for track and get track id 

def search_for_track(track, artist):
    
    # Gets token and builds parts needed to authorize get request
    token = get_token()
    headers = {"Authorization": f"Bearer {token}"}
    endpoint = "https://api.spotify.com/v1/search"
    
    # Builds url to search for specific track
    data = urlencode({"q": f"track:{track} artist:{artist}", "type": 'track'})

    lookup_url = f"{endpoint}?{data}"

    # Gets token and authorizes request
    info = requests.get(lookup_url, headers = headers)
    # Returns track id and headers for furter use or returns 'None' if no track is found
    if info.json()['tracks']['items'] == []: return None, None, None
    else: 
        return (info.json()['tracks']['items'][0]['id']), headers, (info.json()['tracks']['items'][0]['name'])
```

```python
# Get track audio features using track id and convert form json to dataframe    

def get_features(track_id, headers):
    
    fts = requests.get('https://api.spotify.com/v1/audio-features/' + track_id, headers = headers)
    res_df = fts.json()

    # Creates a dataframe of the selected track to add to the exisiting dataframe for the purpose of scaling
    row = pd.DataFrame(
    {'acousticness':[res_df['acousticness']], 
     'danceability':[res_df['danceability']], 
     'energy':[res_df['energy']], 
     'instrumentalness':[res_df['instrumentalness']], 
     'key_num':[res_df['key']], 
     'liveness':[res_df['liveness']], 
     'loudness':[res_df['loudness']], 
     'mode_num':[res_df['mode']],
     'speechiness':[res_df['speechiness']],
     'tempo':[res_df['tempo']],
     'time_signature_num':[res_df['time_signature']],
     'valence':[res_df['valence']]})
    
    # Scales data
    row.loc[row.index, features] = scaler.transform(row)
    
    # Returns row of track's audio features
    return row
```

```python
def predict_genres():
   
    # Prompts the user for track and artist
    while True:
        
        track = input('Please enter track name to find its genre:\n').upper()
        artist = input('Please enter artist name of song to find its genre:\n').upper()
        
        # If data is found in the csv loaded dataset
        if df[(df.artist_name == artist) & (df.track_name == track)].shape == (1,40): 
            result = df[(df.artist_name == artist) & (df.track_name == track)]
            break
        # Looks for track using the spotify api if track is not already in the loaded dataset
        else: 
            track_id, headers, name = search_for_track(track, artist)
            if track_id == None: print('Please enter valid input')
            else: 
                result = get_features(track_id, headers)
                break
    
    # Predicts genre of the song
    f_result = result[features]
    pred = rf.predict(f_result)  
    probs = rf.predict_proba(f_result) 
    
    # Gives the next 2 likeliest genres that the song can be classified into in case spotify would want to classify songs into more than one genre
    genre1prob = np.sort(probs[0])[::-1][0]*100
    genre2 = ''.join(rf.classes_[probs[0].argsort()[::-1][1:2]])
    genre2prob = np.sort(probs[0])[::-1][1]*100
    genre3 = ''.join(rf.classes_[probs[0].argsort()[::-1][2:3]])
    genre3prob = np.sort(probs[0])[::-1][2]*100
    
    print('\nThe genre the song can be classified into is: {0} at {1:.2f}% probaility.'.format(''.join(pred), genre1prob))
    print('The next 2 likeliest genres the song can be classified into are: {0} at {1:.2f}% probability and {2} at {3:.2f}% probability.'.format(genre2, genre2prob, genre3, genre3prob)) 
```

# Similar Songs Recommender Function

Recommends the top k similar songs to the one selected by the user.

```python
def knn(songs_data_vecs, song_query_vec, k):
    
    neighbor_distances_and_indices = []
    
    # Euclidean distance calculation 
    # Goes through all the songs in the dataset
    for index, song_dataset_vec in enumerate(songs_data_vecs):
        
        sum_squared_distance = 0
        # Goes through all the features in a song
        for i in range(len(song_dataset_vec)):
            sum_squared_distance += math.pow(song_dataset_vec[i] - song_query_vec[0][i], 2)
        if sum_squared_distance != 0:
            distance = math.sqrt(sum_squared_distance)
            neighbor_distances_and_indices.append((distance, index))
    
    # Sorts the distances from smallest to largest
    sorted_neighbor_distances_and_indices = sorted(neighbor_distances_and_indices)
    
    # Selects the requested k numbers of results
    k_nearest_distances_and_indices = sorted_neighbor_distances_and_indices[:k]
    
    return k_nearest_distances_and_indices
```

```python
def recommend_songs():
    
    # Prompts the user for track, artist, and number of similar recommendations
    while True: 
        
        k = input(f'Please enter the number of similar songs you would like to get recommended between 1 and {len(df)}:\n')
        try: val = int(k)
        except ValueError: continue
        k = int(k)
        if not (k >= 1 and k <= len(df)): continue
            
        track = input(f'Please enter track name to find its {k} closest similar songs:\n').upper()
        artist = input(f'Please enter artist name of song to find its {k} closest similar songs:\n').upper()
        
        # If data is found in the csv loaded dataset
        if df[(df.artist_name == artist) & (df.track_name == track)].shape == (1,40): 
            result = df[(df.artist_name == artist) & (df.track_name == track)]
            break
        # Looks for track using the spotify api if track is not already in the loaded dataset
        else: 
            track_id, headers, name = search_for_track(track, artist)
        
            if track_id == None: print('Please enter valid input')
            else: 
                result = get_features(track_id, headers)
                break
    
    # Create vector of feature values for requested song
    f_result = result[features]
    song_query_vec = f_result.values.tolist()
    
    # Create vectors of feature values for all songs
    songs_data_vecs = X.values.tolist() 

    recommendation_indices = knn(songs_data_vecs, song_query_vec, k)
     
    # Gets the song recommendations
    song_recommendations = []
    for _, index in recommendation_indices:
        song_recommendations.append((df.iloc[index]['track_name'], df.iloc[index]['artist_name']))

    # Prints the song recommendations
    print(f'\nYour top {k} song recommendations are: \n')
    for i in range(len(song_recommendations)):
        print(f'{i+1}. {song_recommendations[i][0]} by {song_recommendations[i][1]}')
     
```
